```python
# Importando o 'pandas' e a base do 'kaggle'
import pandas as pd
treino = pd.read_csv('train.csv')
```

```python
# Visualizando se foi importado corretamente
treino
```

```python
# Verificando as dimensões(shape) da base
treino.shape
```

```python
# Visualizando as informações da base
treino.info()
```

```python
# Confirmando a quantidade de valores nulos por colunas
treino.isnull().sum().sort_values(ascending=False).head(20)
```

```python
# Também pode ser visto a porcentagem de valores nulos por coluna
(treino.isnull().sum() / treino.shape[0] * 100).sort_values(ascending=False).head(20)
```

```python
# Filtrando as colunas mais graves em relação a valores nulos.
# Especificamente colunas com valores acima de 10%.
remover = treino.columns[treino.isnull().sum() / treino.shape[0] * 100 > 10]
remover
```

```python
# Removendo as colunas que possuem uma maior quantidade de valores nulos.
treino = treino.drop(remover, axis=1)
treino
```

```python
# Substituindo os valores vazios restantes
treino = treino.fillna(-1)
treino.head(3)
```

```python
# Filtrando as colunas numéricas para submete-las aos modelos
col_numerica = treino.dtypes != 'object'
col_numerica
```

```python
# Armazenando na base somente as colunas numéricas
treino = treino.loc[:, col_numerica]
treino.head(3)
```

**Para separar em Treino e Teste**
(https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html)

1- Primeiro modelo:
[Regressão Linear](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html)

2- Segundo modelo:
[Árvore de Regressão](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html)

3- Terceiro modelo:
[KNN](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html)

```python
# Importando o treino e teste
from sklearn.model_selection import train_test_split
```

```python
# Separando o X e o 'y'
X = treino.drop('SalePrice', axis=1)
y = treino.SalePrice
```

```python
# Separando em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42
)
```

```python
# Importando o primeiro modelo
from sklearn.linear_model import LinearRegression
clf_rl = LinearRegression()
```

```python
# Fazendo o fit com os dados de treino
clf_rl = clf_rl.fit(X_train, y_train)
```

```python
# Fazendo a previsão com os dados de test
y_pred_rl = clf_rl.predict(X_test)
```

```python
# Importando o segundo modelo
from sklearn.tree import DecisionTreeRegressor
clf_tree = DecisionTreeRegressor(random_state=42)
```

```python
# Fazendo o fit com os dados de treino
clf_tree = clf_tree.fit(X_train, y_train)
```

```python
# Fazendo a previsão com os dados de test
y_pred_tree = clf_tree.predict(X_test)
```

```python
# Importando o terceiro modelo
from sklearn.neighbors import KNeighborsRegressor
clf_knn = KNeighborsRegressor()
```

```python
# Fazendo o fit com os dados de treino
clf_knn = clf_knn.fit(X_train, y_train)
```

```python
# Fazendo a previsão com os dados de test
y_pred_knn = clf_knn.predict(X_test)
```

**Métricas para avaliar os modelos**

1- Primeira métrica:
[Erro Médio Absoluto](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.mean_absolute_error.html)

2- Segunda métrica:
[Erro Médio Quadrático](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.mean_squared_error.html)

```python
# Importando a primeira e segunda métrica
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
# Imprimindo os resultados
print(mean_absolute_error(y_test, y_pred_rl))
print(mean_squared_error(y_test, y_pred_rl))
print()
print(mean_absolute_error(y_test, y_pred_tree))
print(mean_squared_error(y_test, y_pred_tree))
print()
print(mean_absolute_error(y_test, y_pred_knn))
print(mean_squared_error(y_test, y_pred_knn))
```
